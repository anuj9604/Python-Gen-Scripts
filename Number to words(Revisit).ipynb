```python
def numtowords():
    ones=('', 'One', 'Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine')
    tens_excep=('Ten', 'Eleven', 'Twelve', 'Thirteen', 'Fourteen', 'Fifteen', 'Sixteen', 'Seventeen', 'Eighteen', 'Nineteen')
    tens=('Twenty', 'Thirty', 'Forty', 'Fifty', 'Sixty', 'Seventy', 'Eighty', 'Ninety', 'Hundered')
    suffixes=('', 'Thousand', 'Million', 'Billion')
    
    def process(number, index):
        
        if number=='0':
            return 'Zero'
        
        length=len(number)
        
        if(length>3):
            return False
        
        number= number.zfill(3)
        words=''
        
        hdigit=int(number[0])
        tdigit=int(number[1])
        odigit=int(number[2])
        
        words+='' if hdigit==0 else ones[hdigit]+' '+tens[8]+' '
        
        if(tdigit>1):
            words+=tens[tdigit-2]
            words+=' '
            words+=ones[odigit]
            
        elif(tdigit==1):
            words+= tens_excep[(int(tdigit+odigit)%10)-1]
            
        elif(tdigit==0):
            words+=ones[odigit]+' '
            
        if not len(words)==0:
            words+=' '+suffixes[index]
            
        return words
    
    
    def get_words(number):
        length= len(str(number))
        
        if length>12:
            return 'This program support upto 12 digits numbers only'
        
        count=length//3 if length%3==0 else length//3+1
        copy=count
        words=[]
        
        for i in range(length-1, -1, -3):
            if i==0:
                words.append(process(str(number)[0], copy-count))
            elif i==1:
                words.append(process(str(number)[i-1:i+1], copy-count))
            else:
                words.append(process(str(number)[i-2:i+1], copy-count))
            count -= 1;
            
        final_words=''
        for s in reversed(words):
            temp=s+' '
            final_words+=temp
        
        return final_words
    
    number=input("Enter any number: ")
    print(get_words(number))
```
